# Remove Nulls

### Identify Nulls

```python
total = raw_data.isnull().sum().sort_values(ascending=False)
percent = (raw_data.isnull().sum()/raw_data.isnull().count()).sort_values(ascending=False)
raw_data_missingData = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
display(raw_data_missingData.head(20))
raw_data.isnull().sum().max()
```

###  Remove Rows with Nulls

```python
nullsRemoved = pd.DataFrame()
nullsRemoved = raw_data.copy()
nullsRemoved.dropna(inplace=True)
```

```python
print ("Before Dropping Nulls: ", raw_data.shape)
print ("After Dropping Nulls: ", nullsRemoved.shape)
print ("Rows Removed: ", raw_data.shape[0] - nullsRemoved.shape[0])
print ("Percent Removed: {:.0%}".format(1-(nullsRemoved.shape[0])/raw_data.shape[0]))   
```

# Make Binary Classifier

```python
binClsfr = pd.DataFrame()
binClsfr = nullsRemoved.copy()
binClsfr['binaryRating'] = np.where(binClsfr.rating>=2.5, 1, 0)
```

```python
sns.countplot(x=binClsfr.binaryRating, data=binClsfr)
plt.show()
```
